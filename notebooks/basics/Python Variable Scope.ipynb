Local Scope

def value():
    x = 10
    print(x)
value()

```python
print(x)
```

Global Scope

```python
x = 20  
def display():
    print(x)

display()
print(x)
```

Local vs Global Variable Conflict

```python
value = 50

def update():
    value = 10
    print(value)

update()
print(value)
```

Using the global Keyword




```python
count = 0

def increment():
    global count
    count += 1

increment()
print(count)
```

Enclosing Scope (Nested Functions)

```python
def outer():
    message = "Hello"

    def inner():
        print(message)
    inner()
outer()
```

Using the nonlocal Keyword

```python
def outer():
    count = 0

    def inner():
        nonlocal count
        count += 1
        print(count)

    inner()
    inner()

outer()
```

LEGB Rule (Scope Resolution Order)

Python resolves variables using the LEGB rule:

Local

Enclosing

Global

Built-in

```python
x = "Global"

def outer():
    x = "Enclosing"

    def inner():
        x = "Local"
        print(x)

    inner()

outer()
```

Built-in Scope

```python
print(len([1, 2, 3]))
```

Scope Inside Loops

```python
for i in range(3):
    loop = i

print(loop)
```

Practical Example of Scope Management

```python
total = 100

def calculate():
    total = 50

    def adjust():
        nonlocal total
        total += 10

    adjust()
    return total

print(calculate())
```

```python
print(total)  
```
