Splitting and Joining Strings

```python
sentence = "Python is powerful"
words = sentence.split(" ")
print(words)
```

```python
joined = "-".join(words)
print(joined)
```

String Formatting

```python
name = "Swathiii"
score = 100

print(f"{name} scored {score} marks.")  
```

```python
print("{} scored {} marks.".format(name, score)) 
```

 Iterating Through a String

```python
word = "Swathii"

for char in word:
    print(char)
```
