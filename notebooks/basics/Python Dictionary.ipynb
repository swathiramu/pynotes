 Creating a Dictionary

```python
dict = {}
print(dict)
```

```python
student = {"name": "Alice", "age": 22, "grade": "A"}
print(student)
```

Accessing Dictionary Values

```python
emp = {"id": 186, "name": "Swathi", "role": "Developer"}
print(emp)
```

```python
emp = {"id": 186, "name": "Swathi", "role": "Developer"}
print(emp["name"])
```

```python
emp = {"id": 186, "name": "Swathi", "role": "Developer"}
print(emp.get("role"))
```

Adding and Updating Dictionary Entries

```python
prof = {"username": "admin"}
prof["email"] = "swathiramu28@gmail.com"
prof["username"] = "admin"
print(prof)
```

Removing Items from a Dictionary

```python
data = {"a": 1, "b": 2, "c": 3}
removed = data.pop("b")
del data["a"]
print(data)    
```

```python
print(removed)
```

Dictionary Length and Membership

```python
set = {"theme": "dark", "version": 1.0}
print(len(set))  
```

```python
print("theme" in set) 
```

Iterating Through Dictionaries

```python
user = {"name": "Swathi", "age": 21, "city": "London"}

for key in user:
    print(key, ":", user[key])

for key, value in user.items():
    print(key, "=>", value)
```

Dictionary Methods

```python
data = {"x": 10, "y": 20}
print(data.keys()) 
```

```python
print(data.values())
```

```python
print(data.items()) 
```

Merging Dictionaries

```python
dict1 = {"a": 1, "b": 2}
dict2 = {"c": 3, "d": 4}
merged = dict1 | dict2
print(merged)
```

Nested Dictionaries

```python
employee = {"id": 101,"personal": {"name": "John","email": "john@example.com"}}
print(employee["personal"]["name"])
```

 Dictionary Comprehension

```python
squa = {x: x**2 for x in range(5)}
print(squa) 
```
