Declaring and Assigning Variables

```python
count = 10
price = 99.99
name = "Python"

print(count) 
print(price)   
print(name)    
```

Multiple Variable Assignment

```python
a, b, c = 1, 2, 3
print(a)  
print(b)  
print(c)
```

Reassigning Variables

```python
value = 100
print(value)
```

```python
value = "Reassigned"
print(value)
```

Numeric Literals

```python
integer = 42
print(integer)  
```

```python
float = 3.14
print(float) 
```

```python
complex = 2 + 3j
print(complex)
```

String Literals

```python
single_quote = 'Hello'
print(single_quote)  
```

```python
double_quote = "World"
print(double_quote) 
```

```python
multi_line = """This is a multi-line string"""
print(multi_line)
```

Boolean Literals

```python
is_active = True
is_logged_in = False

print(is_active)    
print(is_logged_in)
```

Special Literal - None

```python
result = None
if result is None:
    print("No value assigned yet")
```

Collection Literals

```python
list_literal = [1, 2, 3]
print(list_literal)
```

```python
tuple_literal = (4, 5, 6)
print(tuple_literal)
```

```python
set_literal = {7, 8, 9}
print(set_literal)
```

```python
dict_literal = {"a": 1, "b": 2}
print(dict_literal)   
```

Binary, Octal, and Hexadecimal Literals

```python
binary = 0b1010
print(binary)
```

```python
octal = 0o12
print(octal)
```

```python
hexadecimal = 0xA
print(hexadecimal)  
```

Using Variables in Expressions

```python
x = 5
y = 3

sum_result = x + y
product_result = x * y
print(sum_result) 
```

```python
print(product_result)
```
