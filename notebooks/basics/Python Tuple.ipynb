Creating a Tuple

```python
empty = ()
print(empty)
```

```python
single= (10)
print(single)
```

```python
num = (1, 2, 3, 4)
print(num)
```

```python
mixed = (10, "Python", True)
print(mixed)
```

Accessing Tuple Elements

```python
colors = ("red", "green", "blue")
print(colors[0])
```

```python
print(colors[-1])
```

```python
print(colors[-2])
```

```python
print(colors[1])
```

Slicing Tuples

```python
values = (0, 1, 2, 3, 4)
print(values[1:4])
```

```python
values = (0, 1, 2, 3, 4)

print(values[0:3])
```

Tuple Packing and Unpacking

```python
person = ("Swathiii", 21, "Engineer")
name, age, profession = person
```

```python
print(name) 
```

```python
print(age) 
```

```python
print(profession)
```

Nested Tuples

```python
data = ((1, 2), (3, 4), (5, 6))
print(data[1])
```

```python
print(data[2])
```

```python
print(data[1][0])
```

```python
print(data[2][1])
```

Tuple Length and Membership

```python
num = (10, 20, 30)
print(len(num))
```

```python
numbers = (10, 20, 30, 78, 90)

print(len(numbers))
```

```python
print(20 in numbers) 
```

```python
print(60 in numbers) 
```

Iterating Through a Tuple

```python
languages = ("Python", "Java", "Go")
for lang in languages:
    print(lang)
```

Converting Between Tuples and Lists

```python
values = (1, 2, 3)

list1 = list(values)
tuple1 = tuple(list1)

print(list1)   
print(tuple1) 
```
