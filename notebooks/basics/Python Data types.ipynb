Python Numbers and Mathematics

Numeric Data Types in Python

```python
int = 10
print(type(int))
```

```python
float = 3.14
print(type(float))
```

```python
complex = 2 + 5j
print(type(complex))
```

Basic Mathematical Operations

```python
a = 12
b = 5
print(a + b) 
```

```python
print(a - b)
```

```python
print(a * b)
```

```python
print(a / b)
```

Floor Division and Modulus

```python
x = 17
y = 5

print(x // y) 
```

```python
print(x % y)
```

```python
base = 2
exponent = 4

res = base ** exponent
print(res)
```

Math Module Basics

```python
import math
print(math.sqrt(25))
```

```python
print(math.ceil(4.3))
```

```python
print(math.floor(4.9))
```

Trigonometric Functions

```python
import math
a = math.radians(30)
print(math.sin(a)) 
```

```python
print(math.cos(a))
```

```python
print(math.tan(a)) 
```

 Rounding Numbers

```python
val = 3.5678
print(round(val)) 
```

```python
print(round(val, 2)) 
```

 Absolute and Sign Functions

```python
num = -25
print(abs(num)) 
```

```python
print(math.copysign(10, -1))
```

Random Number Generation

```python
import random
print(random.randint(1, 30))
```

```python
print(random.uniform(1.0, 5.0))
```

Statistical Operations

```python
import statistics
num = [10, 20, 30, 40, 50]
print(statistics.mean(num))   
```

```python
print(statistics.median(num)) 
```

```python
print(statistics.stdev(num))  
```
