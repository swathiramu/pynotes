Automatic Removal of Duplicates

```python
num = {1, 2, 2, 3, 3, 4, 6, 7, 7, 8 , 8}
print(num) 
```

```python
f = {"apple", "banana"}
f.add("orange")
print(f)
```

Adding Multiple Elements with update()

```python
c = {"red", "green"}
c.update(["blue", "yellow"])

print(c)
```

Removing Elements from a Set

```python
num = {10, 20, 30, 40, 50}
num.remove(20)
print(num)
```

```python
num = {10, 20, 30, 40, 50}
num.discard(50)
print(num)
```

```python
num = {10, 20, 30, 40, 50}
num.discard(10)
print(num)
```

```python
num = {10, 20, 30, 40, 50}
num.discard(30)
print(num)
```

 Set Length and Membership

```python
data = {5, 10, 15}
print(len(data))  
```

```python
data = {5, 10, 15, 78, 56,43}
print(len(data))  
```

```python
print(10 in data) 
```

```python
print(100 not in data)
```

```python
print(10 not in data)
```

 Iterating Through a Set

```python
language = {"Python", "Java", "Go"}
for lang in language:
    print(lang)
```

Mathematical Set Operations

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}
print(A | B)
```

```python
print(A & B)  
```

```python
print(A - B)
```

```python
print(A ^ B)
```

```python
A = {1, 2, 3}
B = {1, 2}

print(B.issubset(A)) 
```

```python
print(A.issuperset(B))
```

```python
print(A.issubset(B)) 
```

Converting Sets to Other Types

```python
set = {1, 2, 3}
listv = list(set)
print(listv) 
```

```python
tuplev = tuple(set)
print(tuplev)
```
