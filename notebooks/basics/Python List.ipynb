Slicing Lists

```python
num = [0, 1, 2, 3, 4, 5]
print(num[1:4])
```

```python
print(num[:3])
```

```python
print(num[3:]) 
```

Modifying List Elements

```python
A = ["pen", "pencil", "eraser"]
A[1] = "marker"

print(A) 
```

```python
A = ["pen", "pencil", "eraser"]
A[2] = "scale"

print(A) 
```

Adding Elements to a List

```python
num = [1, 2, 3]
num.append(4)
print(num)
```

```python
num = [1, 2, 3]
num.insert(1, 10)
print(num)
```

Removing Elements from a List

```python
data = [10, 20, 30, 40]
data.remove(20)
print(data)
```

```python
data = [10, 20, 30, 40]
last = data.pop()
print(data)
```

List Length and Membership

```python
values = [5, 10, 15]
print(len(values)) 
```

```python
print(10 in values)  
```

```python
print(189 not in values)
```

 List Iteration

```python
c = ["New York", "London", "Tokyo"]
for city in c:
    print(city)
```

List Comprehension

```python
A = [x**2 for x in range(5)]
print(A)
```

```python
A = [x**2 for x in range(10)]
print(A)
```
